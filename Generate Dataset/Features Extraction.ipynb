### Different port

```python
def diff_port(url):
    p = '(?:http.*://)?(?P<host>[^:/ ]+).?(?P<port>[0-9]*).*'

    m = re.search(p,url)
    if m.group('port') :
        return 1
    else :
        return 0
diff_port('http://www.abc.com:123/test')
```
